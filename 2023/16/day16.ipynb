```python
with open(R'input/day16.txt','r') as t:
    puzzleInput = t.read().splitlines()
puzzleInput = [list(row) for row in puzzleInput]

from collections import deque

def part1(grid: list)->int:
    start = [(0, -1, 'E')] # direction light is headed to (NESW)
    q = deque(start)
    seen = set()
    mirror1 = { # /
                "N":"E",
                "E":"N",
                "S":"W",
                "W":"S"
    }

    mirror2 = { # \
                "N":"W",
                "W":"N",
                "S":"E",
                "E":"S"
    }

    while q:
        # print(q)
        i, j, dir = q.popleft()
        
        # Move in that direction    
        match dir:
            case 'E':
                j += 1
            case 'W':
                j -= 1
            case 'N':
                i -= 1
            case 'S':
                i += 1

        if i < 0 or i >= len(grid) or j < 0 or j >= len(grid[0]):
            continue
        else:
            cur = grid[i][j]
            if cur == "." or (cur == '-' and dir in ['E','W']) or (cur == '|' and dir in ['N','S']):
                if (i,j,dir) not in seen:
                    q.append((i,j,dir))
                    seen.add((i,j,dir))
            elif cur == "/":
                if (i,j,mirror1[dir]) not in seen:
                    q.append((i,j,mirror1[dir]))
                    seen.add((i,j,mirror1[dir]))
            elif cur == "\\":
                if (i,j,mirror2[dir]) not in seen:
                    q.append((i,j,mirror2[dir]))
                    seen.add((i,j,mirror2[dir]))
            elif cur == "-" and dir in ['N','S']:
                if (i,j,'E') not in seen:
                    q.append((i,j,'E'))
                    seen.add((i,j,'E'))
                if (i,j,'W') not in seen:
                    q.append((i,j,'W'))
                    seen.add((i,j,'W'))
            elif cur == "|" and dir in ['W','E']:
                if (i,j,'N') not in seen:
                    q.append((i,j,'N'))
                    seen.add((i,j,'N'))
                if (i,j,'S') not in seen:
                    q.append((i,j,'S'))
                    seen.add((i,j,'S'))
    return len({(i,j) for (i,j,_) in seen})

part1(puzzleInput)
```

```python
def part2(grid: list, i0,j0,dir)->int:
    start = [(i0, j0, dir)] # direction light is headed to (NESW)
    q = deque(start)
    seen = set()
    mirror1 = { # /
                "N":"E",
                "E":"N",
                "S":"W",
                "W":"S"
    }
    mirror2 = { # \
                "N":"W",
                "W":"N",
                "S":"E",
                "E":"S"
    }
    while q:
        # print(q)
        i, j, dir = q.popleft()
        
        # Move in that direction    
        match dir:
            case 'E':
                j += 1
            case 'W':
                j -= 1
            case 'N':
                i -= 1
            case 'S':
                i += 1

        if i < 0 or i >= len(grid) or j < 0 or j >= len(grid[0]):
            continue
        else:
            cur = grid[i][j]
            if cur == "." or (cur == '-' and dir in ['E','W']) or (cur == '|' and dir in ['N','S']):
                if (i,j,dir) not in seen:
                    q.append((i,j,dir))
                    seen.add((i,j,dir))
            elif cur == "/":
                if (i,j,mirror1[dir]) not in seen:
                    q.append((i,j,mirror1[dir]))
                    seen.add((i,j,mirror1[dir]))
            elif cur == "\\":
                if (i,j,mirror2[dir]) not in seen:
                    q.append((i,j,mirror2[dir]))
                    seen.add((i,j,mirror2[dir]))
            elif cur == "-" and dir in ['N','S']:
                if (i,j,'E') not in seen:
                    q.append((i,j,'E'))
                    seen.add((i,j,'E'))
                if (i,j,'W') not in seen:
                    q.append((i,j,'W'))
                    seen.add((i,j,'W'))
            elif cur == "|" and dir in ['W','E']:
                if (i,j,'N') not in seen:
                    q.append((i,j,'N'))
                    seen.add((i,j,'N'))
                if (i,j,'S') not in seen:
                    q.append((i,j,'S'))
                    seen.add((i,j,'S'))
    return len({(i,j) for (i,j,_) in seen})

curMax = 0
for j in range(len(puzzleInput)):
    top2bot = part2(puzzleInput, -1, j, 'S')
    bot2top = part2(puzzleInput, len(puzzleInput), j, 'N')
    l2r = part2(puzzleInput, j, -1, 'E')
    r2l = part2(puzzleInput,j,len(puzzleInput), 'W')
    result = max(top2bot, bot2top, l2r, r2l)
    if result > curMax:
        curMax = result

curMax
```
